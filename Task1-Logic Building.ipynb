```python
#User Login Check 
username="admin"
password="1234"

name_input=input("Enter a username: ")
password_input=input("Enter a Password: ")

if (name_input==username and password_input==password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

total_pass_count=0
total_fail_count=0

for m in marks:
    if m>=50:
        total_pass_count+=1
    else:
        total_fail_count+=1

print("Student Marks Analysis")
print("Total Students pass: ",total_pass_count)
print("Total Students fail: ",total_fail_count)
```

```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

result=[]    #Create a empty list to store final result

for n in names:
    r= n.strip()  #To remove extra space 
    r= r.lower()  #To convert into lowercase

    result.append(r)  #To append the clean and standardized data into the result list 
print("Clean and Standardized data")
print(result)
```

```python
#Message Length Analyzer 

messages = ["Hi", "Welcome to the platform", "OK"]

for m in messages:
    print(f'{m}: {len(m)}',end=' ')

    if(len(m) >10):
        print("'| Alert!! This message has more than 10 Characters")
    else:
        print("| This is OK ")
    
```

```python
#Error Message 

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count=0

for l in logs: 
    if l=='ERROR':
        error_count+=1

print(f"Total Error Count Entries: {error_count}")
    
```
