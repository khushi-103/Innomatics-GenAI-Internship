# Problem 1:- Unique words in a sentences
Problem Statement :
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
sentence=input("Enter Sentence:")

word_list=sentence.split()
unique_word_set=set(word_list)

print(f"Unique Word Count:{len(unique_word_set)}")
print(f"Unique Words:{unique_word_set}")
```

# Problem 2:- Highest Salary from Employee Data
Problem Statement :
Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees={
    "Riya":80000,
    "Omkar":75000,
    "Shradha":85000,
    "Yash":90000
}

highest_sal_employee=""
highest_salary=0

for name,salary in employees.items():
    if salary> highest_salary:
        highest_salary=salary
        highest_sal_employee=name

print(f"Highest Salary: {highest_sal_employee} - {highest_salary}")
```

# Problem 3:- Find Maximum and Minimum Values
Problem Statement :
Find the maximum and minimum values from a list of numbers.


```python
numbers=[45,32,56,36,33,78,12,6,90,2]
min_value=numbers[0]
max_value=numbers[0]

for num in numbers:
    if num<min_value:
        min_value=num
    if num>max_value:
        max_value=num
print(f"Min: {min_value}")
print(f"Max: {max_value}")
```

# Problem 4:- Count Products Above a Price Threshold
Problem Statement :
Count how many products have a price greater than 1000.


```python
prices=[550,346,2000,5000,3400,500,5000]

threshold_value=1000
count=0

for i in prices:
    if i>threshold_value:
        count+=1

print(f"Product above 1000:- {count}")
```

# Problem 5:- Calculate Attendance Percentage
Problem Statement :
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance=['P','P','A','P','A','A','P']
            
count=0
total_days=len(attendance)

for i in attendance:
    if i=="P":
        count+=1
        
attendance_percentage=(count/total_days)*100
print(f"Attendance Percentage: {round(attendance_percentage,1)}")
```

# Problem 7:- Count Character Frequency
Problem Statement :
Count how many times each character appears in a given string.


```python
input_string=input("Enter a string:")

dictionary={}

for i in input_string:
    if i in dictionary:
        dictionary[i]+=1
    else:
        dictionary[i]=1
print(dictionary)
```

# Problem 10:- Calculate Average Marks
Problem Statement : 
Calculate the average marks of a student from a list of marks.


```python
student_marks=[45,35,65,34,76,80]

total_marks=sum(student_marks)
avg_marks=total_marks/len(student_marks)

print("Average Marks :",round(avg_marks,2))
```
