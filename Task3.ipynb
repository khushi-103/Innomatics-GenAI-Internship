# Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
top_employee=[]
max_score=max(employees.values())

for name,score in employees.items():
    if score==max_score:
        top_employee.append(name)
print(f'Top Performers Eligible for Bonus: {', '.join(top_employee)} (Score: {max_score})')
#Here, to display the top employees, I used the join() method to merge all the names in the list into a single string, separated by a comma and a space.
```

# Problem Statement 2: Search Query Keyword Analysis

```python
# Importing regular expression module to clean punctuation from text
import re 

user_input=input("Enter your input: ")

# Converting the entire sentence to lowercase
user_input=user_input.lower()

''Removing all punctuation marks using regex
  [^\w\s] means remove everything except letters, numbers, and spaces'''
user_input=re.sub(r'[^\w\s]','',user_input)

# Splitting the cleaned sentence into individual words
words=user_input.split()

# Create a dictionary to store word frequencies
word_count={}

# Create a dictionary to store words that appear more than once
final_result={}

# Loop through each word in the list
for word in words:
    if word in word_count:
        word_count[word]+=1
    else:
        word_count[word]=1

# Loop through dictionary items to filter repeated words
for word,count in word_count.items():
    if count>1:
        final_result[word]=count

print(final_result)
```

In the above code, first I convert the input to lowercase to avoid case-sensitive mismatches. Then I remove punctuation using regular expressions. After splitting the sentence into words, I use a dictionary to count the frequency of each word. Finally, I filter out and display only those words that appear more than once.

# Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_reading=[]

for hour , value in enumerate(sensor_readings):
    if value%2==0:
        # Store (hour_index, reading_value) in valid_reading list
         valid_reading.append((hour,value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_reading)
```

# Problem Statement 4:- Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
total_emails = len(emails)

for email in emails:
    # Split email at '@' to get the domain part
    domain = email.split("@")[1]
    # Increase count of that domain and if not present start from 0
    domain_count[domain] = domain_count.get(domain, 0) + 1

#Loop through each domain and count and calculate its percentage 
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# Problem Statement 5:- Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000,5000,5400]

# Calculate average
average = sum(sales) / len(sales)

# Calculate 30% above average
threshold = average * 1.30

# Check for spikes
for i in range(len(sales)):
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```

# Problem Statement 6:- Duplicate User ID Detection


```python
'''# Import Counter class from collections module
# Counter helps to count occurrences of elements in a list'''

from collections import Counter 
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

## Count how many times each user ID appears in the list
total_count=Counter(user_ids)

for user_id,count in total_count.items():
    if count>1:
        print (f"{user_id} --> {count} times")
```

First, I use Counter from the collections module to count how many times each user ID appears. Then I iterate through the dictionary of counts. If a user ID appears more than once, I print it as a duplicate along with the number of occurrences.
