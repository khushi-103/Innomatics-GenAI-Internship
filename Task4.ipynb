# Problem 1: Mobile Recharge Validation System

``` 
Problem Statement:  Create a function to validate a mobile recharge request.
Rules:
    Recharge amount must be ≥ ₹50
    Valid plans: 199, 299, 399, 599
    User should be allowed to retry if invalid
```


````python
def recharge_validation(amount):
    if amount<50:
        print("Minimum recharge amount is ₹50.")
        return False
        
    elif amount not in [199, 299, 399, 599]
        print("Invalid plan selected.")
        print("Available plans are: 199, 299, 399, 599.")
        return False
    else:
        return True

while True:
    print("-------------------------------------------------------------")
    recharge_amount=int(input("Enter your Recharge Amount:"))
    result=recharge_validation(recharge_amount)

    if result==True:
        print("Recharge successful!")
        break;
    if result==False:
        print("Please try Again!")
````

# Problem 2: Inventory Reorder Alert System
```
Problem Statement :

Given a dictionary of products and stock quantities:
    Print Reorder Alert if stock < 15
    Print Stock OK otherwise
```


````python
products_info={
    "Laptop": 10,
    "Headphones": 18,
    "Smartwatch": 7,
    "Tablet": 22,
    "Camera": 14}

def stock_monitor(product,stock):
    if stock<15:
        print(f"{product}:-Reorder Alert!! (Current Stock: {stock})")
    else:
        print(f"{product}:-Stock OK!! (Current Stock: {stock})")

for product,quantity in products_info.items():
       print('--------------------------------------------------------')
       stock_monitor(product,quantity)

    
````

# Problem 3: Student Result Processing System
```
Problem Statement :-

Create a function that:
    Accepts a list of student marks
    Calculates average marks
    Prints Pass if average ≥ 50 else Fail
```

````python
def marks_analysis(marks_list):
    total=0
    
    for mark in marks_list:
        total+=mark
    average=total/len(marks_list)
    print(f"Average Marks : {round(average,2)}")

    if average>=50:
        return "Pass"
    else:
        return "Fail"

marks_list=[]

total_subject=int(input("Enter the total number of subjects:"))

for i in range(total_subject):
    marks=float(input(f"Enter the number of marks for subject {i+1} : "))
    marks_list.append(marks)
print("--------------------------------------------------------------")

result=marks_analysis(marks_list)
print(f"Final Result : {result}")
        
````

# Problem 4:- Cab Fare Estimator with Retry Option
```
Problem Statement : 
    Create a cab fare calculator:

    Base fare = ₹50
    ₹12 per km
    Peak hour → 25% extra
    Allow retry using while loop
```

````python
def fare_calculator(distance,hour):
    base_fare=50
    per_km_rate=12
    
    total_fare=base_fare+(distance*per_km_rate)

    if hour=='yes':
        total_fare=total_fare*1.25

    return total_fare


while True:
    distance=int(input("Enter the distance:"))
    if distance<=0:
        print("Distance must be greater than 0")
        continue
        
    hour=input("Enter if it is peak hour or nor(yes/no): ").lower()

    result=fare_calculator(distance, hour)

    print(f"Total fare of the ride : {result}")
    print("-------------------------------------------------------------------------------")

    retry_choice=input("Do you want to calculate the fare price again (yes/no) : ").lower()
    if retry_choice=='yes':
        continue
    elif retry_choice=='no':
        break
    else:
        print("Invalid Choice")
````

# Problem 5: Employee Attendance Eligibility Checker
```
Problem Statement : 

Create a function that:
    Accepts attendance list ("P", "A")
    Calculates attendance percentage
    Returns Eligible if ≥ 75% else Not Eligible
```

````python
def calculate_attendance(attendance):
    present_count=0
    for i in attendance:
        if i=='P':
            present_count+=1
    percentage=(present_count/len(attendance))*100
    print(f"Attendance Percentage: {round(percentage,2)}%")

    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
    
attendance_list=[]
total_days=int(input("Enter the number of days : "))
for i in range(total_days):
    attendance=input(f"Enter attendance (P/A) for day{i+1} :").upper()
    attendance_list.append(attendance)

result=calculate_attendance(attendance_list)
print(result)
````

# Problem 6: Password Strength Checker
```
Problem Statement : Create a function to check password strength.
Rules:
    Minimum length = 8
    Must contain at least one digit
    Must contain at least one special character (@#$)
```

````python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password (Minimum length should be 8 characters)"

    has_digit = False
    has_special = False
    special_characters = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"

user_password = input("Enter your password: ")
result = check_password_strength(user_password)

print(result)
````
