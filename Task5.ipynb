# Problem 1: Social Media – Post Engagement Analyzer
```
Problem Statement : Create a function to analyze engagement on a social media post.
    
Rules :
    Calculate total likes
    If Total Likes ≥ 1000 → Viral Post
    Otherwise → Normal Engagement
```


````python
def post_analyses(data_list):
    total_count=0

    for data in data_list:
        total_count+=data

    print(f"Total Likes : {total_count}")
    if total_count>=1000:
        return "Viral Post"
    else:
        return "Normal Engagement"


likes_data=[45,101,500,370,800]
result=post_analyses(likes_data)
print(f"Post Status : {result}")
````

# Problem 3: Agriculture – Rainfall Adequacy Checker
```
Problem Statement : Create a function that analyzes seasonal rainfall data.
Rules :
    Calculate average rainfall
    Average ≥ required level → Adequate Rainfall
    Otherwise → Inadequate Rainfall
```


````python
def rainfall_data_analysis(data,level):
    total_rainfall=0
    for i in data:
        total_rainfall+=i
    average_rainfall=total_rainfall/len(data)
    print(f'Average Rainfall : {average_rainfall}')

    if average_rainfall>=level:
        print("Rainfall Status : Adequate Rainfall")
    else:
        print("Rainfall Status : Inadequate Rainfall")

rainfall_data = [70, 75, 68, 80, 67]
required_level = 70
    
rainfall_data_analysis(rainfall_data,required_level)
````

# Problem 4: Social Media – Duplicate Account Detection
```
Problem Statement : Create a function to detect duplicate usernames in a system.
Rules : 
    If duplicates exist → Flag accounts
    Else → Accounts are unique
```

````python
def duplicate_detection(username):
    unique_username=set(username)

    if len(unique_username)!=len(username):
        return "Yes"
    else:
        return "No"


username_data=['alice','bob','alice','charlie','charlie']

result=duplicate_detection(username_data)
print(f'Duplicate Account Found: {result}')
                   
````

# Problem 6: Agriculture – Premium Crop Price Filter
```
Problem Statement : Create a function that filters crop prices above a premium threshold.
Rules:
    Price > ₹2000 → Premium Crop
    Else → Regular Crop
```

````python
def filter_prices(price):
    premium_price=[]
    for p in price:
       if p>2000:
           premium_price.append(p)

    return premium_price
        


prices=[5600,3000,1500,2300,1800,1000]

result=filter_prices(prices)
print(f"Premium Crops:{result}")
````

# Problem 9: E-Learning – Student Attendance Eligibility System
```
Problem Statement : Create a function to calculate attendance percentage and check exam eligibility.
Rules : 
    Attendance ≥ 75% → Eligible
    Attendance < 75% → Not Eligible
```

````python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    
    
    for status in attendance_list:
        if status == 1:   # 1 means present
            attended_classes += 1
    
    attendance_percentage = (attended_classes / total_classes) * 100 #To calculate the percentage of attendance 
    
    print(f"Attendance Percentage: {attendance_percentage}")
    
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


#Sample attendance data here 1 = Present and  0 = Absent
attendance_data = [1, 1, 1, 1, 0]  

check_exam_eligibility(attendance_data)
````

# Problem 10: Smart Electricity Bill Analyzer
```
Problem Statement: You are building a smart electricity billing system for residential users. Create a function that calculates the electricity bill 
and classifies usage level based on units consumed.

Billing Rules : 
    First 100 units → ₹3 per unit
    Next 100 units (101–200) → ₹5 per unit
    Above 200 units → ₹7 per unit

Usage Classification : 
    Total bill < ₹500 → Low Usage
    ₹500 – ₹1500 → Moderate Usage
    ₹1500 → High Usage
```

````python
def calculate_electricity_bill(units):
    bill = 0

    #Slab calculation
    
    #If consumption is within first 100 units
    if units <= 100:
        bill = units * 3
        
    #If consumption is between 101 and 200 units
    elif units <= 200:
        #First 100 units at ₹3
        #Remaining units at ₹5
        bill = (100 * 3) + (units - 100) * 5

    #If consumption is above 200 units
    else:
        #First 100 units at ₹3
        #Next 100 units at ₹5
        #Remaining units at ₹7
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    #Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status



units_consumed=int(input("Please Enter the units consumed: "))
total_bill, usage_status = calculate_electricity_bill(units_consumed)


print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
````
