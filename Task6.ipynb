# Problem Statement 1:Smart Parking Lot Management System
```
Design a function to manage a smart parking lot.
The system should:
    Accept vehicle entry and exit logs
    Calculate total parked vehicles
    Identify peak parking usage
    Alert if parking exceeds capacity
```


````python
def smart_parking(capacity,logs):
    current_vehicles=0
    peak_vehicles=0  #Highest number of vehicles parked 

    for log in logs:
        if log=="IN":
            current_vehicles+=1
        elif log=="OUT":
            if current_vehicles>0:
                current_vehicles-=1
    if current_vehicles>peak_vehicles:
        peak_vehicles=current_vehicles
    print(f"Currently Parked Vehicles:{current_vehicles}")

    if current_vehicles>capacity:
        
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity=50
logs=["IN","OUT","IN","IN","IN","OUT","OUT","IN"]

smart_parking(capacity,logs)
````

# Problem Statement 2:Online Food Delivery Time Estimator
```
Create a function that estimates delivery time based on:
    Distance (km)
    Weather condition
    Traffic level
Apply delays dynamically and display final ETA.
```

````python
def time_estimator(distance,traffic,weather):
    speed=20

    base_time=(distance/speed)*60
    
    #Traffic delay
    if traffic=="low":
        base_time+=5
    elif traffic=="medium":
        base_time+=10
    elif traffic=="high":
        base_time+=15

    #Weather delay
    if weather=="clear":
        base_time+=0
    elif weather=="cloudy":
        base_time+=5
    elif weather=="rainy":
        base_time+=10
    elif weather=='stormy':
        base_time+=20

    print(f"Estimated Delivery Time: {round(base_time)} minutes")

distance=float(input("Distance (km): "))
traffic=input("Traffic Level: ").lower()
weather=input("Weather: ").lower()

time_estimator(distance,traffic,weather)
````

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
```
Build a function that analyzes seat booking data and:
    Calculates occupancy percentage
    Determines if show is Housefull
    Suggests opening additional shows
```

````python
def seat_occupancy_analyzer(total_seats, booked_seats):
    
    total_booked = len(booked_seats)
    
    # Calculate occupancy percentage
    occupancy_percentage = (total_booked / total_seats) * 100
    
    # Determine show status
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    # Suggest additional show based on occupancy_percentage
    if occupancy_percentage >= 90:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Additional Show Needed"
    
    
    print(f"Occupancy: {occupancy_percentage:.0f}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")


# Sample Input
total_seats = 200
booked_seats = [1] * 100   

seat_occupancy_analyzer(total_seats, booked_seats)
````

# Problem Statement 4:Cloud Server Load Classification System
```
Create a function to classify server load based on CPU usage readings.
Rules:
    Average CPU < 50% → Normal
    50%–80% → Warning
    80% → Critical
```

````python
def classify_server_load(cpu_readings):
    
    total = 0
    for reading in cpu_readings:
        total += reading
    
   
    average_cpu = total / len(cpu_readings)
    
   
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    
    print(f"Average CPU Load: {average_cpu:.0f}%")
    print(f"Server Status: {status}")

cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
````

# Problem Statement 5:Smart Classroom Resource Usage Monitor
```
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.
```

````python
def resource_usage_monitor(resource_usage):
    overused_resources = []

    # Define overuse threshold in hours
    threshold = 8

    # Checking each resource
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")



resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_usage_monitor(resource_data)
````

# Problem Statement 6: Online Event Registration Capacity Controller
```
Create a function that manages event registrations by:
    Tracking registrations
    Preventing overbooking
    Triggering waitlist mode
```

````python
def manage_event_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Process each registration
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")



event_capacity = 100
total_registrations = 105

manage_event_registrations(event_capacity, total_registrations)
````
