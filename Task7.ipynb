# Problem 1: Smart Light Controller
```
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
```

````python
class SmartLight:
    def __init__(self,name,action):
        self.name=name
        self.action=action
    def action_function(self):
        if self.action=='ON':
            print(f"{self.name} is ON")
        elif self.action=='OFF':
            print(f"{self.name} is OFF")
        else:
            print("Invalid Action")

name=input("Enter Light Name:")
action=input("Enter action:").upper()

#create a object of class
obj=SmartLight(name,action)

#Method call
obj.action_function()
        
````

# Problem 2: Employee ID Card System
```
Problem Statement : 
Create a class that stores employee name, ID, and department and displays ID card details.
```

````python
class student_ID:
    def __init__(self,name,ID,department):
        self.name=name
        self.ID=ID
        self.department=department

    def display_info(self):
        print('------------------------------------------------')
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.ID}")
        print(f"Department: {self.department}")

name=input("Employee Name: ")
emp_id=input("Employee ID: ")
department=input("Department: ")

#Creating Object
s=student_ID(name,emp_id,department)

#Displaying ID card
s.display_info()
````

# Problem 3: Mobile Contact Record
```
Problem Statement : 
Create a class to store contact name and phone number and display contact information.
```

````python
class contact:
    def __init__(self,name,number):
        self.name=name
        self.number=number
    def show_info(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone No: {self.number}")

name=input("Contact Name: ")
phone_no=int(input("Phone Number: "))

c=contact(name,phone_no)
c.show_info()

             
````

# Problem 4: Product Price Tag Generator
```
Problem Statement : 
Create a class that stores product name and price and prints a formatted price tag.
```

````python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price

    def display(self):
        print("\nProduct:",self.name)
        print(f"price: ₹{self.price}")
product_name=input("Product Name:")
price=int(input("Price:"))

#creating object
p=Product(product_name,price)

p.display()
````

# Problem 5: Movie Rating Display System
```
Problem Statement :
Create a class that stores movie name and rating and displays movie details.
```

````python
class Movies:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating

    def display_info(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating}/5")

movie_name=input("Movie name:")
rating=float(input("Rating:"))

m=Movies(movie_name,rating)
m.display_info()
                 
````

# Problem 6: Delivery Address Manager
```
Problem Statement :
Create a class that stores customer name and delivery address and prints delivery details.
```

````python
class Customer:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def display_result(self):
        print("\nDelivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

name=input("Customer Name:")
address=input("Address:")

c=Customer(name,address)
c.display_result()
````
